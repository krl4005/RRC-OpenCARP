```python
import matplotlib.pyplot as plt
import numpy as np 

def get_data(filename):
    data = np.loadtxt(filename)

    t_data = []
    v_data = []
    i_ion = []
    for i in list(range(0, len(data))):
        t_data.append(data[i][0])
        v_data.append(data[i][1])
        i_ion.append(data[i][2])
        
    return(t_data, v_data, i_ion)
```

```python
# GRANDI MODEL
# bench --imp Grandi --stim-file stims_grandi.dat -Ograndi_0.001 --dur 9000 --dt "0.001"
fig, axs = plt.subplots(1, figsize = (18, 5), constrained_layout = True)

# dt = 0.1
t_grandi_0, v_grandi_0, i_grandi_0 = get_data('grandi_0.1.txt')
axs.plot(t_grandi_0, v_grandi_0, label = 'dt = 0.1')

# dt = 0.01
t_grandi_1, v_grandi_1, i_grandi_1 = get_data('grandi_0.01.txt')
axs.plot(t_grandi_1, v_grandi_1, label = 'dt = 0.01')

# dt = 0.001
t_grandi_2, v_grandi_2, i_grandi_2 = get_data('grandi_0.001.txt')
axs.plot(t_grandi_2, v_grandi_2, label = 'dt = 0.001')

axs.set_title('Grandi Model | stims = [0.45, 0.5, 0.55] | RRC = 0.45')
axs.legend()
```

```python
# OHARA MODEL
#  bench --imp OHara --stim-file stims_ohara.dat -Oohara --dur 18000
fig, axs = plt.subplots(1, figsize = (18, 5), constrained_layout = True)
t_ohara, v_ohara, i_ohara = get_data('ohara.txt')
axs.plot(t_ohara, v_ohara)
axs.set_title('OHara Model | stims = [0.6, 0.7, 0.73] | RRC = 0.7')
```
